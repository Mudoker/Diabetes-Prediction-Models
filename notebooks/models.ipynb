```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd
import sys

sys.path.append("../")
# Import user-defined packages
from scripts.models.linear_regression import LinearRegression as MyLinearModel
```

```python
# df = pd.read_csv("../data/processed/data_train_processed.csv")
df = pd.read_csv("../data/processed/data_train_processed.csv")
df.describe().round(2)
```

```python
# Drop the 'Status' column for features and assign it to X
X = df.drop(columns=["Status"])

# Assign the 'Status' column to y as the target variable
y = df["Status"]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# Creating and training the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluation
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)
mse = mean_squared_error(y_test, model.predict(X_test))

print("Training R^2 Score:", train_score)
print("Testing R^2 Score:", test_score)
print("Mean Squared Error:", mse)
print("Coefficient: ", model.coef_)
```

```python
# Self defined linear regression model
linear_regression_model = MyLinearModel()

linear_regression_model.fit(X=X_train, y=y_train)

predictions = linear_regression_model.predict(X_test)

mse = linear_regression_model.mean_squared_error(y_true=y_test, y_pred=predictions)
print("Mean Squared Error:", mse)
print("Coefficient: ", linear_regression_model.theta)
```
